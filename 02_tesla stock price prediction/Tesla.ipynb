```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from matplotlib import style
style.use("fivethirtyeight")
```

```python
#Read the csv file into a data frame
df = pd.read_csv("Tesla.csv+-+Tesla.csv.csv")
```

```python
#making 2 new columns that will be used for making predictions
df["HL_Perc"] = (df["High"]-df["Low"])/df["Low"] * 100
df["CO_Perc"] = (df["Close"]-df["Open"]) / df["Open"] * 100
```

```python
dates  = np.array(df["Date"])
dates_check = dates[-30:]
dates = dates[:-30]
```

```python
df = df[["HL_Perc", "CO_Perc", "Adj Close", "Volume"]]
```

```python
#Define the label column
df["PriceNextMonth"] = df["Adj Close"].shift(-30)
```

```python
df.tail()
```

```python
#Make the features and label the arrays
X = np.array(df.drop(["PriceNextMonth"], 1))
X  = preprocessing.scale(X)
X_Check = X[-30:]
X = X[:-30]
df.dropna(inplace=True)
y = np.array(df["PriceNextMonth"])
```

```python
#Divide the data set into training data and testing data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Defining the production model
model = RandomForestRegressor()
```

```python
#Fitting the model using training data
model.fit(X_train, y_train)
```

```python
#Calculate the confidence value by applying the model to the testing data
conf = model.score(X_test, y_test)
print(conf)
```

```python
#Fit the model again using the whole data set
model.fit(X, y)
```

```python
predictions  = model.predict(X_Check)
```

```python
#Creating the final data frame containing Dates, ClosePrices, and Forecast Values
actual = pd.DataFrame(dates, columns=["Date"])
actual["ClosePrice"] = df["Adj Close"]
actual["Forecast"] = np.nan
actual.set_index("Date", inplace=True)
forecast = pd.DataFrame(dates_check, columns=["Date"])
forecast["Forecast"] = predictions
forecast["ClosePrice"] = np.nan
forecast.set_index("Date", inplace=True)
var = [actual, forecast]
result = pd.concat(var) #the final data frame
```

```python
#Plotting the final results!
result.plot(figsize = (20, 10), linewidth = 2.1)
plt.legend(loc = 2, prop = {'size' : 15})
plt.xlabel('Date')
plt.ylabel('Price')
```
